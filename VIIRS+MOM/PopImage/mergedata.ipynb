A notebook to test the ways to merge data with pandas framework

```python
import pandas as pd
```

```python
hwrf = "testdata/Final_Attributes_2022122606HWRF+20221225DFO+20221225VIIRSUpdated.csv"
popcount = "processing/Final_Attributes_2022122606HWRF+20221225DFO+20221225VIIRSPopCount.csv"
allpopcount = "popcount_2020.csv"
```

```python
hwrf_df = pd.read_csv(hwrf)
pop_df = pd.read_csv(popcount)
allpop_df = pd.read_csv(allpopcount)
```

```python
# drop totalpop column
pop_df = pop_df.drop(['totalpop'], axis=1)
```

```python
joindpop_df = allpop_df.set_index('pfaf_id').join(pop_df.set_index('pfaf_id'))
```

```python
joindpop_df.head()
```

```python
merged_df = hwrf_df.set_index('pfaf_id').join(joindpop_df)
```

```python
merged_df.head()
```

```python
# need to move column position
popcols = joindpop_df.columns.tolist()
popcols
```

```python
hwrfcols = hwrf_df.set_index('pfaf_id').columns.tolist()
hwrfcols
```

```python
viirspos=hwrfcols.index("VIIRSTotal_Score")
```

```python
newcols = hwrfcols[:viirspos+1] + popcols + hwrfcols[viirspos+1:]
newcols
```

```python
len(newcols)
```

```python
len(merged_df.columns.tolist())
```

```python
reordered_df = merged_df[newcols]
```

```python
reordered_df.head()
```

```python
reordered_df.to_csv("megredhwrf.csv")
```
